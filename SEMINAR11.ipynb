# Даны две функции:
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# Найти точки пересечения
# Построить графики функций в одной системе координат
# Построить график общей функции


```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
```

```python
a=plot(f,g)
```

```python
z = f-g
a = plot(z)
```

```python
x_array = solve(z)
[[x_el.n().round(2),f.subs(x, x_el).n().round(2)]for x_el in x_array]
```
